```python
# Binomial Law
data = np.random.binomial(1,0.7,size = 100000)
sns.distplot(data)
```

```python
# Samples mean with Binomial law
data_split = np.array_split(data,1000)
mean_split = [np.mean(a) for a in data_split]
mu = np.mean(mean_split)
std = np.std(mean_split)
norm = np.random.normal(mu,std,1000)

sns.distplot(mean_split)
sns.distplot(norm,hist=False)
plt.show()
```

```python
# Gamma law
data = np.random.gamma(4,2, size = 1000000)
sns.distplot(data)
```

```python
# Samples mean of Gamma law
data_split = np.array_split(data,1000)
mean_split = [np.mean(a) for a in data_split]
mu = np.mean(mean_split)
std = np.std(mean_split)
norm = np.random.normal(mu,std,1000)

sns.distplot(mean_split)
sns.distplot(norm,hist=False)
plt.show()

```

```python
# Poisson law
data = np.random.poisson(lam = 4, size = 100000)
sns.distplot(data)
```

```python
# Samples mean of Poisson law
data_split = np.array_split(data,1000)
mean_split = [np.mean(a) for a in data_split]
mu = np.mean(mean_split)
std = np.std(mean_split)
norm = np.random.normal(mu,std,1000)

sns.distplot(mean_split)
sns.distplot(norm,hist=False)
plt.show()
```

to understand what kind of random phenomena these laws describe.

1. **Binomial law**: for repetitions of experiments with binary results and on 
sums the number of positive results, for example the number of faces obtained 
after tossing a coin a number of times.

2. **Uniform law**: the complete equiprobable randomness.

3. **Bernouilli's Law**: a random experiment with a binary result, for example a coin toss and see if it's a coin toss.

4. **Poisson's law**: number of events occurring over a given period of time, for example the number of people arriving at a bus stop after a period of time equal to scale (the parameter of the law).

5. **Geometrical Law**: prints with discount in a Bernoulli urn with a p proportion of red balls.

6. **Exponential law**: probability that an event occurs after a time equal to the scale parameter. It is used in reliability checks for
evaluate the life of a component
